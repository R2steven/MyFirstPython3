```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        #return array
        threeSum = []

        #sort array to use two pointer (mvt)
        nums.sort()

        #length of array
        n = len(nums)        

        for i in range(n-2): #ends at third from last entry
            #skip if next entry is copy
            if i>0 and nums[i]==nums[i-1]: 
                i+=1
                continue

            j = i+1 # i+1!=i
            k = n-1

            #start two pointer on range(i,n)
            while (j<k):
                sums = nums[i]+nums[j]+nums[k]
                
                #if 3Sum, add to answer
                if sums==0:
                    box = [nums[i],nums[j],nums[k]]
                    if box not in threeSum:
                        threeSum.append(box)
                    #jump past duplicates
                    while(j<k and nums[j]==nums[j+1]):
                        j+=1
                    while(j<k and nums[k]==nums[k-1]):
                        k-=1
                    j+=1 #move pointers
                    k-=1
                # if sum too small, move j up
                elif sums<0:
                    j+=1
                # if sum too large, move k down
                elif sums>0:
                    k-=1
                

                

        return threeSum
```
