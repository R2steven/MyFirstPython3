```python
class Node():

    def __init__(self, vertex = None, next = None, prev = None,*, arrL = list, dictL = dict ):
        #print(vertex, next, prev, arrL, dictL)
        if vertex is not None:
            self.single_init(vertex, next, prev)
        elif arrL[0] is not None:
            self.arrL_init(arrL)
        elif dictL[0] is not None:
            self.dictL_init(dictL)

    def set_vert(self, vertex):
        self._vertex = vertex
    
    def set_next(self, next):
        self._next = next

    def set_prev(self, prev):
        self._prev = prev


    def single_init(self, vertex, next, prev):
        self.set_vert(vertex)
        self.set_next(next)
        self.set_prev(prev)

    def arrL_init(self, arrL):
        self.set_vert(arrL[0])
        if len(arrL) == 2:
            self.set_next(Node(arrL[1]))
        else:
            self.set_next(Node(arrL = arrL[1:]))
        self._prev = None

    def dictL_init(self, dictL):
        pass


    def get_list(self, arrL = []):
        arrL.append(self)
        if(self._next==None):
            return arrL
        return(self._next.get_list(arrL))

    def findT(self, target, path):
        path.append(self)
        try:
            if self._vertex == target._vertex: return path 
            return self._next.findT(target, path)
            
        except:
            return None


    def __info__(self):
        print(f'name: {self._vertex} {type(self._vertex)}' , f'next: {self._next} {type(self._next)}', f'prev: {self._prev} {type(self._prev)}')
```
